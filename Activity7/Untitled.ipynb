```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("C:/Users/vuyis/Downloads/dataset - 2020-09-24.csv")
```

```python
plt.scatter(df['Wins'],df['Losses'])
plt.title('scatter plot for Wins and Losses')
plt.xlabel('Wins')
plt.ylabel('Losses')
plt.grid(True)
plt.show()
```

```python
data= np.random.normal(loc=20.0,scale=5.0,size=365)
plt.hist(data,bins=20,color='blue',alpha=0.7)
plt.title("Histogram for Wins and Losses")
plt.xlabel('Wins')
plt.ylabel('Losses')
plt.show()
```

```python
data = np.random.normal(loc=20.0, scale=5.0, size=365)
intervals = np.arange(0, 51, 5) 
counts, _ = np.histogram(data, bins=intervals)
plt.bar(intervals[:-1], counts, width=4, color='blue', alpha=0.7)
plt.title("Bar graph for Wins and Losses")
plt.xlabel('Wins')
plt.ylabel('Losses')
plt.show()
```

```python
data = np.random.normal(loc=20.0, scale=5.0, size=365)
threshold = 20.0
wins = len(data[data > threshold])
losses = len(data[data <= threshold])
labels = ['Wins', 'Losses']
sizes = [wins, losses]
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie chart for Wins and Losses')
plt.show()
```
